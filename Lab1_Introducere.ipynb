In exemplu de mai jos se modeleaza o lista de elemente. Spre deosebire de C/C++ crearea si utilizarea listelor este mult mai simpla. Functia ***append()*** adauga elemente in lista. Pentru a accesa elemnte din lista folositi operatorul **[ ]**.

```
lista = []
lista.append(7)
lista.append('Acesta este un string')
lista.append(8.9)
print(lista[0])
print(lista[1])
print(lista[2])
```

Pentru parcurgerea unei liste se foloseste ***for***. O lista se poate parcurge direct pe elementele din aceasta sau folosind un index dupa cum arata exemplul de mai jos. Folosind ***range()*** se specifica intervalul ce va fi parcurs. Functia ***len()*** returneaza lungimea listei. 

```
#parcurgea directa 
for element in lista:
  print(element)
#parcurgerea prin index
for i in range(len(lista)):
  print(lista[i])
```

In exemplu de mai sus, varibila *i* va itera prin intervaul determinat de range(len(lista)) adica print multimea: 0, 1 si 2 (deoarece lista are 3 elemente). La modul general, daca se doreste un control mai mare asupra intervalului parcurs, functiei range i se vor specifica 3 parametrii: start (de unde incepe intervalul), stop (unde se opreste), step (pasul cu care se intereaza). Acesta sintaxa este echivalenta cu sintaxa clasica a unui for in C: for(i=start; i<stop; i=i+step). Priviti exemplul de mai jos.

```
lista = []
for i in range(10): #deci 0, 1, 2,...,9
  lista.append(i)
#acum se intereaza din 2 in 2 de la 0 pana a 6
for i in range(0, 7, 2): #deci 0, 2, 4, 6
  print(lista[i])
```

***TODO:*** Scrieti o functie care primeste ca si argumente doua liste: o lista de numere intregi si o lista de valori cuprinse intre 0 si 1. Functia returneaza media ponderata a numerelor din prima lista cu ponderile din cea de a doua lista. Exemplu: lista1={2, 3, 5}, lista2={0.2, 0.3, 0.5} rezultat: 2x0.2+3x0.3+5x0.5=3.8.

```
def mediaponderata(numere, pondere):
  suma=float(0.0)
  for i in range(len(numere)):
    suma+=numere[i]*pondere[i]
  return suma
x = input("Numarul de elemente ")
lista = [];
listaponderi = [];
for i in range(0,int(x)):
  lista.append(float(input("Numarul " + str(i) + " :")))
  listaponderi.append(float(input("Ponderea " + str(i) + " :")))
print ("Media ponderata este: " + str(mediaponderata(lista,listaponderi)))
```

Fie o lista de coordonate x={1,2,3,4} si o lista de coordonate y={7,8,9,10}. Se doreste afisarea punctelor de coordonate (x[i], y[i]).

```
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [7, 8, 9, 10]
#traseaza puntele
plt.scatter(x, y)
#titlul graficului
plt.title('Primul grafic in Python')
#denumire axe
plt.xlabel('Axa x')
plt.ylabel('Axa y')
plt.show()
```

Se observa ca axele sunt scalate automat. Axa x se desfasoara de la 1 la 4 iar axa y de la 7 la 10. Pentru setarea manuala a limitei unei axe se foloseste functia ***plt.axis***.

```
xmin = 1
xmax = 10
ymin = 1
ymax = 10
plt.axis([xmin, xmax, ymin, ymax])

plt.scatter(x, y)
plt.title('Al doilea grafic in Python')
plt.xlabel('Axa x')
plt.ylabel('Axa y')
plt.show()
```

Folosind functia ***np.linspace(start, stop, num)*** se genereaza un numar de **num** numere din intervalul **start**, **stop** egal departate intre ele. In exemplul de mai jos se afiseaza suprafata $z=sin(xy)$.

```
import plotly.graph_objects as go
import numpy as np
x=np.linspace(-2,2, 100)
y=np.linspace(-2,2, 100)
X,Y=np.meshgrid(x,y)
Z=np.sin(X*Y) 

fig = go.Figure(data=[go.Surface(z=Z)])
fig.show()
```

Pentru a afisa in 3D o multime discreta de puncte se foloseste functia ***go.Scatter3d*** din aceeasi librarie. In exemplu de mai jos, am parametrizat sinusul generat anterior.

```
import plotly.graph_objects as go
import numpy as np

t = np.linspace(0, 10, 50)
#x, y, z = np.sin(t), np.sin(t**2), t
x, y, z = np.cos(t), np.sin(t), t
fig = go.Figure(data=[go.Scatter3d(x=x, y=y, z=z,
                                   mode='markers')])
fig.show()
```
